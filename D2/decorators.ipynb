## 1. Funções são objetos de primeira classe

Em Python, funções podem ser passadas como argumento, atribuídas a variáveis e retornadas de outras funções.

```python
def saudacao(nome: str) -> str:
    return f"Olá, {nome}!"

# atribuindo a variável
dizer_oi = saudacao
print(dizer_oi("Antonio"))

# passando como argumento
def aplicar(func, valor: str) -> str:
    return func(valor)

print(aplicar(saudacao, "mundo"))
```

## 2. Funções internas e retorno de função (closure)

Uma função pode definir e retornar outra função. Isso é a base dos decorators.

```python
def pai():
    print("pai() chamado")

    def filho():
        print("filho() chamado")

    return filho  # retorna a função, não a chama

minha_func = pai()
minha_func()
```

## 3. O que é um decorator

Um decorator é uma função que recebe outra função, adiciona comportamento e retorna uma nova função — sem modificar o original.

```python
def meu_decorator(func):
    def wrapper():
        print("antes de chamar a função")
        func()
        print("depois de chamar a função")
    return wrapper

def diz_oi():
    print("oi!")

# aplicando manualmente
diz_oi = meu_decorator(diz_oi)
diz_oi()
```

## 4. Sintaxe com `@`

`@meu_decorator` acima de uma `def` é açúcar sintático para `func = meu_decorator(func)`.

```python
@meu_decorator
def diz_tchau():
    print("tchau!")

diz_tchau()
```

## 5. Decorando funções com argumentos

O `wrapper` precisa aceitar e repassar os argumentos da função original com `*args` e `**kwargs`.

```python
def log_chamada(func):
    def wrapper(*args, **kwargs):
        print(f"chamando {func.__name__} com args={args} kwargs={kwargs}")
        resultado = func(*args, **kwargs)
        print(f"{func.__name__} retornou {resultado!r}")
        return resultado
    return wrapper

@log_chamada
def soma(a: int, b: int) -> int:
    return a + b

@log_chamada
def saudacao(nome: str, formal: bool = False) -> str:
    prefixo = "Sr(a)." if formal else ""
    return f"Olá, {prefixo} {nome}!".strip()

soma(3, 4)
saudacao("Antonio", formal=True)
```

## 6. Stacking de decorators

Múltiplos decorators aplicados de baixo para cima: `@a` sobre `@b` equivale a `a(b(func))`.

```python
def negrito(func):
    def wrapper(*args, **kwargs):
        return f"<b>{func(*args, **kwargs)}</b>"
    return wrapper

def italico(func):
    def wrapper(*args, **kwargs):
        return f"<i>{func(*args, **kwargs)}</i>"
    return wrapper

@negrito
@italico
def formata(texto: str) -> str:
    return texto

# equivale a: negrito(italico(formata))("Python")
print(formata("Python"))
```
