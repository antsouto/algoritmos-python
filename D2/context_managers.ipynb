# Context Managers and Using Python's with Statement
https://realpython.com/ref/stdlib/contextlib/

```python
# Quick toy example

from contextlib import contextmanager

@contextmanager
def context():
    print("Entering")
    yield
    print("Exiting")

with context():
    print("Inside")
```

```python
# Using @contextmanager to create a function-based context manager

@contextmanager
def writable_file():
    try:
        print("Open file for writing...")
        file = open("hello.txt", mode="w", encoding="utf-8")
        yield file
    finally:
        print("Closing file...")
        file.close()

with writable_file() as file:
    file.write("Hello, World!")
```

```python
# Automatically closing a file using the closing() context manager

from contextlib import closing
from urllib.request import urlopen

with closing(urlopen("http://www.example.com")) as page:
    for line in page:
        print(line)
```

```python
# Supress specific exceptions with the supress() context manager

from contextlib import suppress

with suppress(FileNotFoundError):
    open("non_existing_file.txt")
```

```python
# Real-World Example
# Manage multiple resources

from contextlib import ExitStack
files = ["file1.txt", "file2.txt"]

with ExitStack() as stack:
    file_objects = [
        stack.enter_context(open(fname, "w")) for fname in files
    ]
    for file in file_objects:
        file.write("Hello, World!")
```
