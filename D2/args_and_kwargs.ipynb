### Using the Python args Variable in Function Definitions

```python
def my_sum(*integers):
    result = 0
    for x in integers:
        result += x
    return result

print(my_sum(1, 2, 3))
```

All that matters here is that you use the unpacking operator (*).

Bear in mind that the iterable object you’ll get using the unpacking operator * is not a list but a tuple. A tuple is similar to a list in that they both support slicing and iteration. However, tuples are very different in at least one aspect: lists are mutable, while tuples are not.

### Using the Python kwargs Variable in Function Definitions

```python
def concatenate(**words):
    result = ""
    # iterating over the Python kwargs dictionaru
    for arg in words.values():
        result += arg
    return result

print(concatenate(a="Real", b="Python", c="Is", d="Great", e="!"))
```

### Ordering Arguments in a Function

The correct order for your parameters is:

1. Standard arguments
2. *args arguments
1. **kwargs arguments

```python
def my_function(a, b, *args, **kwargs):
    pass
```

```python
def my_function(a, b, **kwargs, *args):
    pass
```

### Unpacking With the Asterisk Operators: * & **
In short, the unpacking operators are operators that unpack the values from iterable objects in Python. The single asterisk operator * can be used on any iterable that Python provides, while the double asterisk operator ** can only be used on dictionaries.

```python
my_list = [1, 2, 3]
print(my_list)
print(*my_list)
```

```python
list1 = [1, 2, 3]
list2 = [4, 5]
list3 = [6, 7, 8, 9]
print(my_sum(*list1, *list2, *list3))
```

```python
### Split a line in three parts
a, *b, c = list3
print(a, b, c)
```

```python
my_merged_list = list1 + list2 + list3
print(my_merged_list)
my_merged_list = [*list1, *list2, *list3]
print(my_merged_list)
```

### `+` vs unpacking para mesclar listas

Ambos produzem o mesmo resultado quando todos os operandos são `list`. A diferença real:

- **`+` aceita só listas** — misturar tipos causa `TypeError`.
- **Unpacking aceita qualquer iterável** (tuple, generator, range…) — uma única alocação, sem listas intermediárias.

Use `[*a, *b, *c]` quando os iteráveis podem ser de tipos diferentes.

```python
lista = [1, 2]
tupla = (3, 4)
gen = (x for x in [5, 6])

print([*lista, *tupla, *gen])  # ✓ funciona com tipos mistos

try:
    print(lista + tupla)       # ✗ TypeError
except TypeError as e:
    print(f"TypeError: {e}")
```

### `**` para mesclar dicts

O ganho aqui é diferente: **não mutabilidade**.

| Abordagem | Muta o original? | Cria novo dict? |
|---|---|---|
| `a.update(b)` | ✓ sim | ✗ não |
| `{**a, **b}` | ✗ não | ✓ sim |
| `a \| b` (3.9+) | ✗ não | ✓ sim |

Vantagens adicionais do unpacking:
- Mistura unpacking com chaves inline: `{**defaults, "timeout": 30}` — padrão comum para configs/overrides.
- `{**a, **b, **c, "k": v}` é mais flexível que encadear `|` quando há muitas fontes ou chaves extras.
- Funciona com qualquer `Mapping`, não só `dict`.

```python
defaults = {"timeout": 10, "retries": 3, "verbose": False}
overrides = {"timeout": 30, "verbose": True}

# update() muta o original
config = defaults.copy()
config.update(overrides)

# unpacking: cria novo dict, defaults intacto
config_unpack = {**defaults, **overrides}

# inline com chave extra
config_extra = {**defaults, **overrides, "version": 2}

print(defaults)       # intacto
print(config_unpack)
print(config_extra)

# Python 3.9+: | para merge simples
config_pipe = defaults | overrides
print(config_pipe)
```

Remember that the * operator works on any iterable object. It can also be used to unpack a string:


```python
a = [*"RealPython"]
print(a)
```
